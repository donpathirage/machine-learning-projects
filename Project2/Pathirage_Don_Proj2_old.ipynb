```python
def cost(X,T,w):  #Cost
    J = np.sum(np.power(T-X@w,2))
    return J
```

```python
def gradient(t,X,w):  #Compute gradient

    grad = 2 * w.T @ X.T @ X - 2*t@X
     
    return grad.T
```

```python
def gradientDescent(X,T,M,rho,epochs):  #Find gradient descent solution
    w = np.random.randn(M+1,1)
    err = np.ones([1,epochs])
    
    for i in range(epochs):
        grad = gradient(T,X,w)
        J = cost(X,T,w)
        print(J)
        w = w - rho*grad
        
#     print(err)
        
#     plt.plot([range(epochs)], err)
#     plt.show()
    
    return w
```

```python
def p(w,x,M):
    y = 0;
    for i in range(M):
        y += w[i]*x**(M-i)
        
    return y
```
